```
df_dropna_cols = df.dropna(axis=1)
display(df_dropna_cols.head())
```

```
df_loc_1 = df.loc[df['Match Year'] >1920]
display(df_loc_1.head())
```

```
df_loc_2 = df.loc[(df['Result'] == 'Lost') ]
display(df_loc_2.head(20))
```

```
df_iloc_1 = df.iloc[:20]
display(df_iloc_1)
```

```
df_iloc_2 = df.iloc[10:20, 2:10]
display(df_iloc_2)
```

```
df_query_1 = df.query("Matches >0")
display(df_query_1.head())
```

```
df_query_1 = df.query(" Matches<1930 ")
display(df_query_1.head())
```

```
df_between_1 = df[df['Match'].between('Australia', 'England')]
display(df_between_1)
```

```
df_mask_1 = df.copy()
df_mask_1['Match Year'] = df_mask_1['Match Year'].mask(df_mask_1['Match Year'] > 1930, 'NOT ELIGIBLE')
display(df_mask_1.head(30))
```

```
df_query_2 = df.query('Result == "Won"')
display(df_query_2.head())
```

```
# Count the occurrences of each result type ('Won', 'Lost', etc.) for each 'Match'
df_groupby = df.groupby('Match')['Result'].value_counts()
display(df_groupby)
```

```
# Find the most frequent result for each 'Match'
df_groupby = df.groupby('Match')['Result'].agg(lambda x: x.mode()[0] if not x.mode().empty else 'No Result')
display(df_groupby)
```
